```python
# if we want to customise seaborn style,we can pass dictionary parameters
# to the set_style() function using axes_style()

x = np.linspace(0,20,50)
y = np.sin(x)
sns.set_style("darkgrid",{'grid.color':'yellow','grid.linestyle':'--','axes.edgecolor':'Red'})
plt.plot(x,y)
```

```python
sns.set_style('whitegrid')
sns.set_context('notebook')
plt.plot(x,y)
```

```python
# to remove axes(top and right axes) spines use sns.despine()
# sns.set_style("ticks",{"axes.spines.left":'False','axes.spines.bottom':'False'})

sns.set_style("ticks")
plt.plot(x,y)
sns.despine()
plt.show()
```

```python
# if i want to use combination of colors use color palette.
# use set_palette()

c = np.sin(x*3)
sns.set() # this theme of sns is applied to all further plots
sns.set_style('ticks')
# set color_palette
sns.set_palette('autumn')
plt.plot(x,np.cos(x))
plt.plot(x,c)

sns.despine()
plt.show()
```

## Seaborn

### Histogram using distance plot

### Histogram represent the data distribution by forming the bins along the range of data 
    # and then drawing bars to show the no. of observations.that fall in each bin.
    

```python
df = sns.load_dataset("iris")
df.head(4)
```

```python
df.species.value_counts()
```

```python
theme# the color theme is this bcoz above we done sns.set() so further graphs show 
plt.hist(df.petal_length)
```

```python
sns.set()
sns.color_palette("icefire") # changed color palette
plt.hist(df.petal_length)
```

```python
# univariate distribution of observation
sns.distplot(df['petal_length'])
plt.show()
```

```python
# dist plot always comes in combination of hist and kde so if we want only one then 
# make false which u dont want
sns.distplot(df['petal_length'],kde=False,bins=[1,2,4,6,8])
plt.show()
```

```python
# kernel density Estimation(kde) is way to estimate probability density function.
# of a continuous random variable
# it is used for non-parametric analysis.

sns.distplot(df['petal_length'],hist=False,bins=[1,2,4,6,8])
plt.show()
```

```python
# if we want combination of dist and kde plot
# univariate distribution

sns.distplot(df['petal_length'])
plt.show()
```

## Univariate, Bivariate and Multivariate

### Bivariate distribution

```python
# bivariate distribution is used to determine the relation betn two variables.
# joint plot is comb of scatter plot + hist plot
sns.jointplot(x='petal_length',y='petal_width',data=df)
plt.show()
```

```python
sns.jointplot(x='petal_length',y='petal_width',data=df,kind='hex')
plt.show()
```

```python
sns.jointplot(x='petal_length',y='petal_width',data=df,kind='reg')
plt.show()
```

```python
sns.jointplot(x='petal_length',y='petal_width',data=df,kind='resid')
plt.show()
```

```python
sns.jointplot(x='petal_length',y='petal_width',data=df,hue='species')
plt.show()
```

pair relationship using pairplot

```python
To plot multiple pairwise distribution in a dataset,we can use pairplot() function.
```

```python
sns.set_style('ticks')
sns.pairplot(df,palette='BrBG')
plt.show()
```

```python
# above the diagonal plots are hist 
# we can change the type of plot

sns.set_style('ticks')
sns.pairplot(df,palette='BrBG',diag_kind='kde')
plt.show()
```

```python
# if we dont want pairplot of whole data
# we can select per need

sns.set_style('ticks')
sns.pairplot(df[['petal_length','sepal_length']],palette='BrBG')
plt.show()
```

```python
# here we can't get both plots bcoz species is categorical data
sns.set_style('ticks')
sns.pairplot(df[['petal_length','species']],palette='BrBG')
plt.show()

# species o/p is not given ,bcoz it is categorical plot.
```

```python
# hue is for categorical data

sns.set_style('ticks')
sns.pairplot(df,palette='BrBG',diag_kind='kde',hue='species')
plt.show()
```

```python
sns.set_style('ticks')
sns.pairplot(df,palette='BrBG',diag_kind='kde',kind='reg')
plt.show()
```
